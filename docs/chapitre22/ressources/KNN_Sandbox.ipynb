```python
import math

def distance_euclidienne(enreg_a, enreg_b, tab_descripteur):
    """
    Paramètres : 
        enreg_a, enreg_b et deux enregistrements de données sous forme de dictionnaires
        tab_descripteur un tableau de descripteurs (certaines clefs des enregistrements)
    Valeur renvoyée :
        distance euclidienne entre enreg_a et enreg_b selon les valeurs des descripteurs de tab_descripteur
    """
    distance_carre = 0
    for descripteur in tab_descripteur:
        distance_carre = distance_carre + (float(enreg_a[descripteur]) - float(enreg_b[descripteur])) ** 2
    return round(math.sqrt(distance_carre), 3)
```

### Test sur le jeu de données iris

```python
table_iris = charger_fichier_entete('./datas/iris.csv')
```

```python
iris_alpha = iris[7]
iris_beta = iris[17]

iris_a_etiqueter = {'longueur_sepale':'5.5', 
                    'largeur_sepale':'3.3', 
                    'longueur_petale':'1.1', 
                    'largeur_petale':'0.2'}


assert distance_euclidienne(iris_alpha, iris_a_etiqueter, ['longueur_sepale', 'longueur_petale']) == 0.64
assert distance_euclidienne(iris_alpha, iris_a_etiqueter, ['longueur_sepale', 'longueur_petale',
                                                          'largeur_petale', 'largeur_sepale']) == 0.648


assert distance_euclidienne(iris_beta, iris_a_etiqueter, ['longueur_sepale', 'longueur_petale']) == 0.5
assert distance_euclidienne(iris_beta, iris_a_etiqueter, ['longueur_sepale', 'longueur_petale',
                                                          'largeur_petale', 'largeur_sepale']) == 0.548
```

```python
def table_distance_nouveau(table, nouveau, distance, etiquette, tab_descripteur):
    tab_distance = []
    for enreg in table:
        tab_distance.append((distance(enreg, nouveau, tab_descripteur), enreg[etiquette]))
    return tab_distance
```

```python
iris_a_etiqueter = {'longueur_sepale':'5.7', 
                    'largeur_sepale':'3.3', 
                    'longueur_petale':'4.7', 
                    'largeur_petale':'0.2'}
tab_distance = table_distance_nouveau(table_iris, iris_a_etiqueter, distance_euclidienne, 'espece',
                                      ['longueur_sepale', 'longueur_petale'])
```

```python
tab_distance[:5]
```

```python
def trier_puis_extraire(tab_distance, k):
    tab_tri = sorted(tab_distance)
    k_voisins = [tab_tri[i][1] for i in range(k)]
    return k_voisins
```

```python
k_voisins = trier_puis_extraire(tab_distance, 5)
```

```python
k_voisins
```

```python
def element_majoritaire(k_voisins):
    occurence = dict()
    voisin_majoritaire = k_voisins[0]
    for voisin in k_voisins:
        if voisin not in occurence:
            occurence[voisin] = 1
        else:
            occurence[voisin] = occurence[voisin] + 1
        if voisin != voisin_majoritaire and occurence[voisin] > occurence[voisin_majoritaire]:
            voisin_majoritaire = voisin
    return voisin_majoritaire    
```

```python
element_majoritaire(k_voisins)
```

```python
def etiquetage_knn(table, nouveau, etiquette, tab_descripteur, k, distance = distance_euclidienne):
    tab_distance = table_distance_nouveau(table, nouveau, distance, etiquette, tab_descripteur)
    k_voisins = trier_puis_extraire(tab_distance, k)
    voisin_majoritaire = element_majoritaire(k_voisins)
    return voisin_majoritaire
    
```

```python
iris_a_etiqueter = {'longueur_sepale':'5.7', 
                    'largeur_sepale':'3.3', 
                    'longueur_petale':'4.8', 
                    'largeur_petale':'0.2'}
```

```python
for k in range(1, 20):
    print(f"k = {k}, etiquette = {etiquetage_knn(table_iris, iris_a_etiqueter,  'espece', ['longueur_sepale', 'longueur_petale'], k)}")
```
