## Exercice 2

```python
import dis 

code_source = """
a = 5
a = a + 3
"""

dis.dis(code_source)
```

```python
list(dis.get_instructions(code_source))
```

```python
code = compile(code_source, '<string>','exec')
for octet in code.co_code:
    print(octet, end = ' ')
```

## Boucle en assembleur

```python
code_ascii = 32
while code_ascii < 127:
    print(chr(code_ascii), end =' ')
    code_ascii = code_ascii + 1
print()
```
